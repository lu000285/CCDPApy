```python
from CCDPApy.helper import input_path
from CCDPApy import FedBatchParameters, FedBatchCellLine, FedBatchExpriment, FedBatchCellCulture

import pandas as pd
pd.set_option('display.max_columns', 200)
```

```python
merk = FedBatchCellCulture()
path = input_path('Merk_Data.xlsx')
merk.load_data(file=path)
merk_data = merk.get_pre_process_data()
merk.get_cell_line_names()
```

```python
cell_line = FedBatchCellLine(cell_line_name='Merk', data=merk_data)
cl_data = cell_line.get_pre_process_data()
cell_line.get_experiment_names()
```

```python
exp = FedBatchExpriment(cell_line_name='Merk', cell_line_id='C3',
                        data=cl_data, use_conc_after_feed=False, use_feed_conc=False)
exp.in_process()
exp.polynomial()
exp.rolling_window_polynomial()
glc = exp.get_species('glucose')
glc.sp_rate_poly
```

```python
parameter = FedBatchParameters(cell_line_name='Merk',
                             use_concentration_after_feed=False,
                             use_feed_concentration=False,
                             regression_method=['polynomial', 'rolling_window_polynomial'],
                             rolling_polynomial_degree=3,
                             rolling_polynomial_window=6)
parameter
```

```python
merk.perform_data_process(parameters=parameter)
```

```python
merk.save_excel(file_name='merk_data_output.xlsx')
```
