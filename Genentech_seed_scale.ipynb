```python
from CCDPApy import FedBatchParameters
from CCDPApy import cell_culture_pipeline
```

```python
param1 = FedBatchParameters(cell_line_name='80L',
                           use_concentration_after_feed=False,
                           use_feed_concentration=False,
                           regression_method=['polynomial']
                           )
param2 = FedBatchParameters(cell_line_name='400L',
                           use_concentration_after_feed=False,
                           use_feed_concentration=False,
                           regression_method=['polynomial']
                           )
param3 = FedBatchParameters(cell_line_name='2000L',
                           use_concentration_after_feed=False,
                           use_feed_concentration=False,
                           regression_method=['polynomial']
                           )
```

```python
measured_data_file = 'genentech_seed_scale.xlsx'
fed_batch = cell_culture_pipeline(cell_culture_type='fed-batch', file=measured_data_file)
```

```python
fed_batch.perform_data_process([param1, param2, param3])
```
