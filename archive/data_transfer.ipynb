```python
from CCDPApy import CellLine    # used to aggregate all cell lines and experiments data
from CCDPApy import bioprocess_pipeline
from CCDPApy.BioProcess.BioProcess import BioProcess
```

```python
input_files = ['04-A1.xlsx', '04-B1.xlsx', '04-C1.xlsx', 'GS_SF13.xlsx']
key = {'use_feed_conc': True,
       'use_conc_after_feed': False,
       'spc_list': ['Glucose','Glutamine','Lactate'],
       'polyreg': True, 
       'rollreg': True, 'rollreg_order': 3, 'rollreg_window': 6,
        }

cl_line = CellLine()

for input in input_files:
        bio_process = bioprocess_pipeline(input_file_name=input, **key)
        cl_line.add_bio_process(bio_process=bio_process)

cl_line.disp_cell_lines()
```

```python
gs1 = cl_line.get_cell_line(cl_name='GS CHOZN Clone 23')
A1 = gs1['CHOMFKBrx04-A1']
B1 = gs1['CHOMFKBrx04-B1']
C1 = gs1['CHOMFKBrx04-C1']
gs2 = cl_line.get_cell_line(cl_name='GS')
```

```python
plot_list = ['Glucose','Lactate','Glutamine']
fig = A1.plot(spc_list=plot_list, profile='all', method='all', file_name='04-A1.png')
fig = B1.plot(spc_list=plot_list, profile='all', method='all', file_name='04-B1.png')
fig = C1.plot(spc_list=plot_list, profile='all', method='all', file_name='04-C1.png')
```

```python
plot_list = plot_list = ['Glucose','Lactate','Glutamine']
file_name = '04'
fig = cl_line.plot_exps(cell_line='GS CHOZN Clone 23', spc_list=plot_list, file_name=file_name)
```

```python
plot_list = plot_list = ['Glucose','Lactate','Glutamine']
file_name = 'GS'
fig = cl_line.plot_exps(cell_line='GS', spc_list=plot_list, file_name=file_name)
```
